```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Создаём и выводим датасет из 10 строк, предсказание изменения цены
base_area = np.array([60, 70, 80, 80, 100, 65, 75, 85, 95, 105])
base_rooms = np.array([1, 2, 2, 3, 3, 1, 2, 2, 3, 3])
base_price = np.array([6.1, 7.3, 8.1, 9.2, 10.0, 6.5, 6.8, 8.5, 9.6, 10.4])

# Доработка: генерируем еще 290 записей,чтобы наполнить датасет
n_samples = 290
area = np.random.choice(base_area, size=n_samples) + np.random.normal(0, 5, n_samples)
rooms = np.random.choice(base_rooms, size=n_samples)
price = area * 0.1 + rooms * 0.5 + np.random.normal(0, 0.3, n_samples)

# Объединяем с данными датасета
df = pd.DataFrame({
    'Площадь': np.concatenate([base_area, area]),
    'Количество комнат': np.concatenate([base_rooms, rooms]),
    'Цена': np.concatenate([base_price, price])
})

print("Первые 25 строк датасета:")
print(df.head(25).to_markdown(index=False))
#записываем данные в csv файл
df.to_csv('price.csv', index=False)
print("\nДатасет сохранен в price.csv")

# #Подготовка данных, выдлаю признаки (х) и переменную(у), также разделяем данные на обучающую и тестовую выборку 70/30
X = df[['Площадь', 'Количество комнат']]
y = df['Цена']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Нормализация данных для KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# инициализация моделей, выбрала 5: лнейная регрессия, дерево решений, случайный лес, ближайшие соседи, градиентный бустинг
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42),
    'K-Neighbors': KNeighborsRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42)
}

# Оптимизация гиперпараметров: подбираем лучшие параметры для каждой модели
params = {
    'Decision Tree': {'max_depth': [2, 3, 5]},
    'Random Forest': {'n_estimators': [50, 100]},
    'K-Neighbors': {'n_neighbors': [3, 5, 7]},
    'Gradient Boosting': {'n_estimators': [50, 100], 'learning_rate': [0.01, 0.1]}
}

# обучение и оценка
#список для хранения резуьтата
results = []
for name, model in models.items():
    if name in params:
        # Оптимизация гиперпараметров, перебираем все возможные комбинации по заданным параметрам
        gs = GridSearchCV(model, params[name], cv=3, scoring='r2')
        gs.fit(X_train_scaled if name == 'K-Neighbors' else X_train, y_train)
        best_model = gs.best_estimator_
    else:
        best_model = model.fit(X_train, y_train)

    # Предсказания
    if name == 'K-Neighbors':
        y_pred = best_model.predict(X_test_scaled)
    else:
        y_pred = best_model.predict(X_test)

    #считаем метрики, название модели, среднеквадратичная ошибка, средняя абсолютная ошибка, коэфф детерминации, параметры
    metrics = {
        'Model': name,
        'MAE': mean_absolute_error(y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'R²': r2_score(y_test, y_pred),
        'Params': gs.best_params_ if name in params else None
    }
    results.append(metrics)

# сравнение моделей, сортируем модели по убыванию и выводим
results_df = pd.DataFrame(results).sort_values('R²', ascending=False)
print("\nСравнение моделей:")
print(results_df.to_markdown(index=False))

# график сравнения моделей
plt.figure(figsize=(12, 6))
#строим линию предсказаний
for model_name in models.keys():
    model = models[model_name]
    if model_name in params:
        gs = GridSearchCV(model, params[model_name], cv=3, scoring='r2')
        gs.fit(X_train_scaled if model_name == 'K-Neighbors' else X_train, y_train)
        model = gs.best_estimator_
    else:
        model.fit(X_train, y_train)

    if model_name == 'K-Neighbors':
        pred = model.predict(X_test_scaled)
    else:
        pred = model.predict(X_test)

    plt.scatter(X_test['Площадь'], y_test, alpha=0.3, label='Данные')
    sorted_idx = np.argsort(X_test['Площадь'])
    plt.plot(X_test['Площадь'].

iloc[sorted_idx], pred[sorted_idx], '--', linewidth=2, label=model_name)

plt.xlabel('Площадь')
plt.ylabel('Цена')
plt.title('Сравнение моделей')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))

# Гистограмма распределения остатко
sns.histplot(residuals, kde=True, bins=15)
plt.axvline(x=0, color='r', linestyle='--')
plt.xlabel('Остатки')
plt.title('гистограмма остатков')

plt.show()

# Выбор лучшей модели
best_model_name = results_df.iloc[0]['Model']
best_model = models[best_model_name]
if best_model_name in params:
    gs = GridSearchCV(best_model, params[best_model_name], cv=3, scoring='r2')
    gs.fit(X_train_scaled if best_model_name == 'K-Neighbors' else X_train, y_train)
    best_model = gs.best_estimator_

print(f"\nЛучшая модель: {best_model_name}")
print(f"Параметры: {results_df.iloc[0]['Params']}")
print(f"R²: {results_df.iloc[0]['R²']:.3f}, RMSE: {results_df.iloc[0]['RMSE']:.3f}")
```
